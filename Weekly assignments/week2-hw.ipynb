**1. Explain with your own words what is a random variable and a random process.**

A random variable is a way to represent the results of a random process.
Similar to in math when you assign a value to a variable (ex. x = 5),
a random variable represents a mapping to the outcome of a random process. For example, if you toss a coin
the possible outcomes are heads or tails. A random variable could represent tails = 1 and heads = 0.
A random process can also be described as a stochastic process and is represented by a collection of operations that usually
are indexed by time and have an outcome that is dependant on probability and isn't deterministic.

**2. Explain with your own words and give an example of the following kind of random processes:**

- (a) Memoryless

    A memoryless process is a process where the probability of acheving a certain outcome in a certain trial does not depend on previous trials.
    For example, flipping a fair coin is a memoryless process because everytime you toss the coin you have a 50% chance of heads or tails regardless of the previous trials.
- (b) Markov

    A Markov random process is a series of events where the probability of subsequent events in the chain is dependant
    exclusively on the state of the immediately preceding event. A game like shoots and ladders is an example of a
    markov random process. The roll of the dice for each move has the same probability every time but the outcome is
    dependant on the players current position on the board at a given time.


- (c) History-dependent

    A history dependant random process is one where the probability of subsequent events is dependant on the previous states.
    History dependant processes are said to have 'memory'. For example when you play a game of poker the probability of
    a certain card appearing on the table will change everytime another card is played so it is advantageous to know or
    remember what cards are on the table.

### Analysis

In all cases simulated in the above exercises the results obtained from the empirical simulation and the analytical formulation
were different. This results because the number of trials was low which makes the values not converge. A well known principal in
computer science dictates that computers are not capable of generating a completely random number because computer hardware is inherently
deterministic. Because of this and other factors, using a computer generated model with a low number of trials will not produce
a result that matches exactly with the analytical formulation. In all cases though the empirical models came close and generally followed
the same shape of the analytical formulation. Both b and c were generally parabolic and a was linearly decaying. With a large number of trials
the empirical formulation and analytical formulation for this distribution will converge.


**4. Using the code from the tutorial, simulate these three scenarios (one scenario per cell, do not combine the three scenarios in the same code). In this case, the restaurant put more works so the service rate varies as follows.**

- (a) For two hours between 9 and 11, the arrival rate is 10 students per 10 minutes. Service rate is 10 students per 10 minutes.

Converted to 60 students per hour with service rate of 60 students per hour

```python
#Code adapted from: https://github.com/sarthak0120/M-M-1-Queue-Simulation
#New libraries needed
import queue #https://docs.python.org/3/library/queue.html
import copy
```

```python
# Input Parameters -- Long one
total_time = 2 #hours
IAT_rate = 60 #elements arriving/hour
ST_rate = 60  #elements served/hour
rho = IAT_rate/ST_rate


# Initialize Parameters
qu = queue.Queue()
curr_process = None
IAT = []
ST = []
AT = []
wait_time = []
server_busy = False
list_wait = []
list_delay = []
num_processes_served=0

IAT.append(0)
i=1
while np.sum(IAT) < total_time*60*60 :
    i = i + 1
    temp = np.random.exponential(1/IAT_rate)*60*60
    IAT.append(int(temp - temp%1))
num_processes = i

# Populate Service-Times (ST) (where ST[i]!=0)
while not len(ST) == num_processes:
    temp = np.random.exponential(1/ST_rate)*60*60
    if not int(temp- temp%1)<1:
        ST.append(int(temp - temp%1))

# Save a copy of ST
ST_copy = copy.deepcopy(ST)

# Get Arrival-Times (AT) from IAT starting at t=0
# and initialize Waiting-Times to 0
for i in range(num_processes):
    if i == 0:
        AT.append(0)
    else:
        AT.append(AT[i-1] + IAT[i])
    wait_time.append(0)

# Simulation of M/M/1 Queue (i represents current time)
queue_size = np.zeros(total_time*60*60)
empty_system=0

for i in range(total_time*60*60):
    queue_size[i] = qu.qsize()
    if server_busy:
        for item in list(qu.queue):
            wait_time[item] = wait_time[item] + 1
        ST[curr_process] = ST[curr_process] - 1
        if ST[curr_process] == 0:
            server_busy = False
            num_processes_served = num_processes_served + 1

    for j in range(num_processes):
        if i== AT[j]:
            qu.put(j)

    if not server_busy and not qu.empty():
        curr_process = qu.get()
        server_busy = True
    #after some transitory
    if not server_busy and qu.empty():
        empty_system = empty_system+1


    sum_wait = 0
    sum_delay = 0
    for i in range(num_processes_served):
        sum_wait = sum_wait + wait_time[i]
        sum_delay = sum_delay + wait_time[i] + ST_copy[i]

    if num_processes_served == 0:
        list_wait.append(0)
        list_delay.append(0)
    else:
        list_wait.append(sum_wait/(num_processes_served*60*60))
        list_delay.append(sum_delay/(num_processes_served*60*60))

###Analytical
# QWT = 1/((1-rho)*ST_rate) - 1/ST_rate
# WT = 1/((1-rho)*ST_rate)


time_hours = np.linspace(0,total_time,total_time*60*60)

plt.figure(figsize=(14,5))
#plt.plot( arrival, 'o',label='Arrival')
#plt.plot( service, 'x',label='Sevice')
plt.plot(time_hours,queue_size, 'ok')

plt.figure(figsize=(14,5))
plt.plot(time_hours, list_wait)
# plt.plot([0, total_time], [QWT, QWT])
plt.ylabel("Avg Wait Times")
plt.show()

plt.figure(figsize=(14,5))
plt.plot(time_hours, list_delay)
# plt.plot([0, total_time], [WT, WT])
plt.ylabel("Avg Delay Times")
plt.show()

plt.figure(figsize=(14,5))
n_bins = np.max(queue_size)
##
plt.hist(queue_size,n_bins.astype(int),density=True)
x = np.arange(0, n_bins+1)
p_analytical = (1-rho)*rho**x
p_analytical[1]= p_analytical[0] + p_analytical[1]
p_analytical[0]= 0
# plt.step(p_analytical,'r')
plt.show()

print('Idle ratio:',empty_system/(total_time*60*60))
print('Empty queue probability:', 1 -  rho)
# print('Wait time in queue',QWT)
# print('Delay time in system',WT)
```

**5. Discussions about the results.**
- What is the impact of a simulation time of 2 hours?

Using 2 hours for the simulation time has the practical implication of making the simulation run in a resonable amount of time on a regular computer.
The main loop contains multiple nested for loops which impacts performance. But in an analysis sense running the simulation for 2 hours provides
a very low sample size which means the results will be more sporadic. When the time is increased the results linearize much more but with this low simulation time
the results follow a similar trend but less closely.

- In the simulations, the 3 periods were also considered independent, i.e. the initial states of the queues were always considered 0. What is the impact of this assumption? (No need to simulate, just arguments)

This impacts the simulation because if the service rate is similar or less than the arrival rate then the queue will not be empty by the time the next period arrives.
This means that the next period cannot be served until the previous queue is empty which in the case of service rate less than arrival rate introduces significant delays.

- Would be this a sound model? Why?

The current model of resetting the queue for each time period is not a sound model to model the entire lunch flow. This is because for the given problem, only 1 of the 3 periods
involves an arrival rate lower than the service rate. Because of this the queue will have a significantamountt of requests still waiting for service (lunch) which will overall impact the wait time
and delay for the rest of the periods.
